```python
import pandas as pd


dfProdutos = pd.read_csv('product_dim.csv')
dfProdutos = dfProdutos[dfProdutos['current_ind'] == 'Y']
display(dfProdutos)
dfProdutos.to_csv('product_dim.csv', index=False)

dfVendas = pd.read_csv('sales_transactions.csv')
display(dfVendas)



```

```python
product_price_dict = dfProdutos.set_index('product_id')['product_price'].to_dict()
print(product_price_dict)

dfVendas['PrecoTotal'] = dfVendas['product_id'].map(product_price_dict) * dfVendas['product_quantity']
dfVendas.to_csv('sales_transactions.csv', index=False)
display(dfVendas)


print(dfVendas.value_counts('PrecoTotal'))
```
